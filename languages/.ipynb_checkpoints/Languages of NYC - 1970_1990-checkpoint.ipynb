# Tokenizing text because I'm really lazy

```python
from textblob import TextBlob
```

```python
excerpt ="それは何年も前だった海の王国で、あなたがアナベル・リーとこの女の名前で知っているかもしれない乙女がそこに住んでいたことは、私に愛し愛されていた"
```

```python
def tokenize(sentence):
    blob = TextBlob(sentence.lower())
    words = [token for token in blob.words if len(token)>1]
    return words
```

```python
tokenize(excerpt)
```
